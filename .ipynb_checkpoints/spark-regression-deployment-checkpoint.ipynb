```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, udf
#from pyspark.sql.types import DoubleType
from pyspark.sql.types import FloatType
import pickle
import json
import pandas as pd
import numpy as np

from sklearn.datasets import load_boston #load boston data
```

```python
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
# Unpickle, pkl file
regmodel = sc.binaryFiles("regmodel.pkl")
model_rdd_data = regmodel.collect()
    
# Load and broadcast python object over spark nodes
#creditcardfrauddetection_model
bostonhousing_model = pickle.loads(model_rdd_data[0][1])
broadcast_bostonhousing_model = sc.broadcast(bostonhousing_model)
```

```python
print(broadcast_bostonhousing_model.value)
```

```python
# Unpickle, pkl file - Scaler Model
scaler_rdd_pkl = sc.binaryFiles("scaling.pkl")
scaler_rdd_data = scaler_rdd_pkl.collect()
    
# Load and broadcast python object over spark nodes
#creditcardfrauddetection_model
scaler_model = pickle.loads(scaler_rdd_data[0][1])
broadcast_scaler = sc.broadcast(scaler_model)
print(broadcast_scaler.value)
```

```python
# Load dataset for prediction
boston_df = load_boston()
```

```python
# Gathering the features of the dataset with column names
X = pd.DataFrame(boston_df.data,columns=boston_df.feature_names)
#Creating a new column for the Target variable
y = boston_df.target
```

```python
# Create feature column list on which model was trained
feature_columns = X.columns.to_list()
print(feature_columns)    
```

## Prediction

### 1. Batch Dataset 

```python
#Scaler model requires data in an array - BATCH -- FULL DATASET
scaler_transform_df = broadcast_scaler.value.transform(np.array(spark_df.collect()))

#Regression model requires data in an array
broadcast_bostonhousing_model.value.predict(scaler_transform_df)
```

### 2. Single Record

```python
#Scaler model requires data in an array - SINGLE RECORD - Need to reshape into a 2D array
scaler_transform_single_df = broadcast_scaler.value.transform(np.array(spark_df.collect()[0]).reshape(1,-1))
#Regression model requires data in an array
broadcast_bostonhousing_model.value.predict(scaler_transform_single_df)
```
